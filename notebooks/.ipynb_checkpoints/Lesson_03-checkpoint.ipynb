```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, classification_report
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/sam16tyagi/Machine-Learning-techniques-in-python/master/logistic%20regression%20dataset-Social_Network_Ads.csv')
```

```python
x = data[['Age', 'EstimatedSalary']]
```

```python
y = data[['Purchased']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=30)
```

```python
model = DecisionTreeClassifier(max_depth=3)
```

```python
model.fit(x_train, y_train)
```

```python
plt.figure(figsize=(10, 5))
plot_tree(model, feature_names = ['Age', 'EstimatedSalary'], class_names=['No', 'Yes'], filled=True)
```

```python
y_pred = model.predict(x_test)
```

```python
confusion_matrix(y_pred, y_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
f1_score(y_pred, y_test)
```

```python
print(classification_report(y_test, y_pred))
```
